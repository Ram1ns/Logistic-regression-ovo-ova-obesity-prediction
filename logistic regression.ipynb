```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = '/ChurnData.csv'
df = pd.read_csv(file)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#df = df[['tenure' , 'age' , 'address' , 'income' , 'ed' , 'equip' , 'churn']]
df['churn'] = df['churn'].astype('int')
df
```

```python
x = np.asarray(df.drop('churn' , axis = 1)) #[['tenure' , 'age' , 'address' , 'income' , 'ed' , 'equip']]
y = np.asarray(df['churn'])
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
print(x_std[ : 5])
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x_std , y , test_size = 0.2 , random_state = 4)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
train = lr.fit(x_train , y_train)
```

```python
y_hat = train.predict(x_test)
print(y_hat)
```

```python
yhat_prob = train.predict_proba(x_test)
print(yhat_prob[ : 5])
```

```python
coef = pd.Series(train.coef_[0] , index = df.columns[:-1])
coef.sort_values().plot(kind = 'barh')
plt.title('coeficient')
plt.xlabel('cofficient values')
#plt.ylabel('coeficient name')
plt.show()
```

```python
from sklearn.metrics import log_loss
log_loss(y_test , yhat_prob)
```
